```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
```

## Initialize signals s, matrix A and their multiplication x

```python
s_1 = lambda t: np.sin(t) + 0.001 * np.random.normal(0,10)
s_2 = lambda t: 2 * (t - np.floor(t)) + 0.002 * np.random.normal(0,10)
s_3 = lambda t: 0.01 * t + 0.001 * np.random.normal(0,10)
ts_1 = np.array([])
ts_2 = np.array([])
ts_3 = np.array([])

time = range(0,1000)
for t in time:
    ts_1 = np.append(ts_1, s_1(t))
    ts_2 = np.append(ts_2, s_2(t))
    ts_3 = np.append(ts_3, s_3(t))

A = np.random.rand(3,3)
f = np.vstack((ts_1, ts_2, ts_3))
x = A.dot(f)
```

```python
plt.figure(figsize=(10,14))
plt.subplot(4,1,1)
plt.title('s_1')
plt.plot(time,ts_1)
plt.subplot(4,1,2)
plt.title('s_2')
plt.plot(time,ts_2)
plt.subplot(4,1,3)
plt.title('s_3')
plt.plot(time,ts_3)
plt.subplot(4,1,4)
plt.title('x')
plt.plot(time,x[0], c = 'tab:red', label = 'x1')
plt.plot(time,x[1], c = 'tab:blue', label = 'x2')
plt.plot(time,x[2], c = 'tab:green', label = 'x3')
plt.legend()
plt.show()
```

## Perform PCA and ICA for signal decomposition

```python
pca = PCA(n_components = 3)
pca.fit(x)
ica = FastICA(n_components = 3, max_iter = 1000)
ica.fit(x)
```

## Results

```python
plt.figure(figsize=(10,12))
plt.subplot(3,1,1)
plt.title('PC1')
plt.plot(time,pca.components_[0])
plt.subplot(3,1,2)
plt.title('PC2')
plt.plot(time,pca.components_[1])
plt.subplot(3,1,3)
plt.title('PC3')
plt.plot(time,pca.components_[2])
plt.show()
```

```python
plt.figure(figsize=(10,12))
plt.subplot(3,1,1)
plt.title('IC1')
plt.plot(time[0:300],ica.components_[0][0:300])
plt.subplot(3,1,2)
plt.title('IC2')
plt.plot(time[0:300],ica.components_[1][0:300])
plt.subplot(3,1,3)
plt.title('IC3')
plt.plot(time[0:300],ica.components_[2][0:300])
plt.show()
```

PCA seems to work for signal decomposition. ICA seems to decompose the modified signal to noise only.
